# 📐 1. Vectores: Los Bloques de Construcción

## ¿Qué es un vector?

Un vector es una **lista ordenada de números** que representa información en el espacio multidimensional.

### Ejemplos del mundo real:
- 👤 **Perfil de cliente**: `[edad, ingresos, gastos]` = `[25, 50000, 30000]`
- 🏠 **Características de casa**: `[m², habitaciones, baños]` = `[120, 3, 2]`
- 🖼️ **Píxel de imagen**: `[rojo, verde, azul]` = `[255, 128, 64]`

---

## 🏗️ Creación de Vectores

```python
# === CREAR VECTORES REPRESENTANDO CLIENTES BANCARIOS ===
# Formato: [edad, ingresos_anuales, gastos_anuales]

cliente_a = np.array([28, 55000, 35000])  # Cliente joven profesional
cliente_b = np.array([35, 70000, 45000])  # Cliente establecido
cliente_c = np.array([42, 85000, 52000])  # Cliente senior

print("🏦 === PERFILES DE CLIENTES BANCARIOS ===")
print(f"Cliente A: {cliente_a} → {cliente_a[0]} años, ${cliente_a[1]:,} ingresos")
print(f"Cliente B: {cliente_b} → {cliente_b[0]} años, ${cliente_b[1]:,} ingresos")
print(f"Cliente C: {cliente_c} → {cliente_c[0]} años, ${cliente_c[1]:,} ingresos")

# === PROPIEDADES BÁSICAS ===
print(f"\n📊 === PROPIEDADES DEL VECTOR ===")
print(f"Forma del vector: {cliente_a.shape}")
print(f"Número de dimensiones: {cliente_a.ndim}")
print(f"Tipo de datos: {cliente_a.dtype}")
print(f"Número de elementos: {cliente_a.size}")

# === DIFERENTES FORMAS DE CREAR VECTORES ===
print(f"\n🛠️ === FORMAS DE CREAR VECTORES ===")

vector_ceros = np.zeros(4)  # Vector de ceros
vector_unos = np.ones(3)    # Vector de unos
vector_rango = np.arange(5, 10)  # Vector secuencial

np.random.seed(42)  # Para reproducibilidad
vector_aleatorio = np.random.normal(100, 15, 5)  # Media=100, std=15

print(f"Vector de ceros: {vector_ceros}")
print(f"Vector de unos: {vector_unos}")
print(f"Vector rango (5-9): {vector_rango}")
print(f"Vector aleatorio: {vector_aleatorio.round(1)}")
```

## ➕ Operaciones Básicas con Vectores

```python
# === SUMA Y RESTA DE VECTORES ===
print("➕ === SUMA DE VECTORES ===")

# La suma combina características de ambos clientes
suma_clientes = cliente_a + cliente_b
promedio_clientes = suma_clientes / 2

print(f"Cliente A: {cliente_a}")
print(f"Cliente B: {cliente_b}")
print(f"Suma:      {suma_clientes}")
print(f"Promedio:  {promedio_clientes.astype(int)}")

print(f"\n💡 Cliente promedio: {promedio_clientes[0]:.0f} años, ${promedio_clientes[1]:,.0f} ingresos")

# === RESTA DE VECTORES ===
print(f"\n➖ === DIFERENCIAS ENTRE CLIENTES ===")
diferencia = cliente_b - cliente_a
print(f"Diferencia B - A: {diferencia}")
print(f"💰 Cliente B gana ${diferencia[1]:,} más que Cliente A")
print(f"🎂 Cliente B es {diferencia[0]} años mayor que Cliente A")

# === MULTIPLICACIÓN POR ESCALAR ===
print(f"\n✖️ === MULTIPLICACIÓN POR ESCALAR ===")
# Simular inflación del 5%
cliente_a_inflacion = cliente_a * 1.05
print(f"Cliente A original: {cliente_a}")
print(f"Con inflación 5%:   {cliente_a_inflacion.astype(int)}")

# Conversión de moneda (USD a EUR, aprox 0.85)
cliente_a_euros = cliente_a.copy().astype(float)
cliente_a_euros[1:] *= 0.85  # Solo convertir valores monetarios
print(f"En euros:           {cliente_a_euros.astype(int)}")
```

## 📏 Magnitud y Normalización

```python
# === MAGNITUD (NORMA) DEL VECTOR ===
print("📏 === MAGNITUD DE VECTORES ===")

# La magnitud mide qué tan "grande" es el vector
magnitud_a = np.linalg.norm(cliente_a)
magnitud_b = np.linalg.norm(cliente_b) 
magnitud_c = np.linalg.norm(cliente_c)

print(f"Magnitud Cliente A: {magnitud_a:,.0f}")
print(f"Magnitud Cliente B: {magnitud_b:,.0f}")
print(f"Magnitud Cliente C: {magnitud_c:,.0f}")

# Cálculo manual de la magnitud para entender mejor
magnitud_manual = np.sqrt(cliente_a[0]**2 + cliente_a[1]**2 + cliente_a[2]**2)
print(f"\n🔢 Cálculo manual Cliente A:")
print(f"√({cliente_a[0]}² + {cliente_a[1]}² + {cliente_a[2]}²) = {magnitud_manual:,.0f}")

# === DIFERENTES TIPOS DE NORMAS ===
print(f"\n📐 === TIPOS DE NORMAS ===")

norma_l1 = np.linalg.norm(cliente_a, ord=1)      # Suma de valores absolutos
norma_l2 = np.linalg.norm(cliente_a, ord=2)      # Euclidiana (default)
norma_inf = np.linalg.norm(cliente_a, ord=np.inf) # Valor máximo

print(f"Norma L1 (Manhattan):  {norma_l1:,.0f}")
print(f"Norma L2 (Euclidiana): {norma_l2:,.0f}")
print(f"Norma L∞ (Máxima):     {norma_inf:,.0f}")

# === NORMALIZACIÓN ===
print(f"\n🎯 === NORMALIZACIÓN DE VECTORES ===")

# Normalizar = crear vector unitario (magnitud = 1)
cliente_a_normalizado = cliente_a / np.linalg.norm(cliente_a)
cliente_b_normalizado = cliente_b / np.linalg.norm(cliente_b)

print(f"Cliente A normalizado: [{cliente_a_normalizado[0]:.3f}, {cliente_a_normalizado[1]:.3f}, {cliente_a_normalizado[2]:.3f}]")
print(f"Nueva magnitud: {np.linalg.norm(cliente_a_normalizado):.10f}")
print(f"💡 Los vectores normalizados siempre tienen magnitud = 1")
```

## 🎯 Producto Punto: La Operación Estrella

```python
# === PRODUCTO PUNTO ===
print("🎯 === PRODUCTO PUNTO ===")

# El producto punto mide qué tan "similares" son dos vectores
producto_ab = np.dot(cliente_a, cliente_b)
producto_ac = np.dot(cliente_a, cliente_c)
producto_bc = np.dot(cliente_b, cliente_c)

print(f"Cliente A · Cliente B = {producto_ab:,}")
print(f"Cliente A · Cliente C = {producto_ac:,}")
print(f"Cliente B · Cliente C = {producto_bc:,}")

# Cálculo manual para entender
producto_manual = (cliente_a[0] * cliente_b[0] + 
                  cliente_a[1] * cliente_b[1] + 
                  cliente_a[2] * cliente_b[2])

print(f"\n🔢 Cálculo manual A·B:")
print(f"{cliente_a[0]}×{cliente_b[0]} + {cliente_a[1]}×{cliente_b[1]} + {cliente_a[2]}×{cliente_b[2]} = {producto_manual:,}")

# === SIMILITUD COSENO ===
print(f"\n🔄 === SIMILITUD COSENO ===")

# Similitud coseno = producto punto de vectores normalizados
sim_ab = np.dot(cliente_a_normalizado, cliente_b_normalizado)
sim_ac = np.dot(cliente_a / np.linalg.norm(cliente_a), cliente_c / np.linalg.norm(cliente_c))

print(f"Similitud A-B: {sim_ab:.3f}")
print(f"Similitud A-C: {sim_ac:.3f}")

# Convertir a ángulo para mejor interpretación
angulo_ab = np.degrees(np.arccos(np.clip(sim_ab, -1, 1)))
angulo_ac = np.degrees(np.arccos(np.clip(sim_ac, -1, 1)))

print(f"\n📐 Ángulos entre vectores:")
print(f"Cliente A-B: {angulo_ab:.1f}°")
print(f"Cliente A-C: {angulo_ac:.1f}°")

# === INTERPRETACIÓN PARA NEGOCIO ===
print(f"\n💼 === APLICACIÓN EN MARKETING ===")

def interpretar_similitud(similitud, nombre1, nombre2):
    if similitud > 0.9:
        return f"✅ {nombre1} y {nombre2} son muy similares - usar misma estrategia"
    elif similitud > 0.7:
        return f"🟡 {nombre1} y {nombre2} son similares - estrategias relacionadas"
    elif similitud > 0.5:
        return f"🟠 {nombre1} y {nombre2} son algo similares - enfoques diferentes"
    else:
        return f"🔴 {nombre1} y {nombre2} son muy diferentes - estrategias distintas"

print(interpretar_similitud(sim_ab, "Cliente A", "Cliente B"))
print(interpretar_similitud(sim_ac, "Cliente A", "Cliente C"))
```

## 📊 Visualización de Vectores 2D

```python
# === VISUALIZACIÓN DE VECTORES ===
print("📊 === VISUALIZACIÓN DE VECTORES 2D ===")

# Crear vectores 2D más simples para visualización
v1 = np.array([3, 4])    # Vector ejemplo 1
v2 = np.array([1, 2])    # Vector ejemplo 2
v3 = v1 + v2             # Suma vectorial

print(f"Vector 1: {v1}")
print(f"Vector 2: {v2}")
print(f"Suma:     {v3}")

# Crear visualización
plt.figure(figsize=(10, 8))

# Dibujar vectores desde el origen
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.008, label=f'Vector 1: {v1}', alpha=0.8)

plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.008, label=f'Vector 2: {v2}', alpha=0.8)

plt.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.01, label=f'Suma: {v3}', alpha=0.9)

# Dibujar la regla del paralelogramo
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, 
           color='blue', alpha=0.4, width=0.005, linestyle='--')

plt.quiver(v2[0], v2[1], v1[0], v1[1], angles='xy', scale_units='xy', scale=1, 
           color='red', alpha=0.4, width=0.005, linestyle='--')

# Configuración del gráfico
plt.xlim(-0.5, 6)
plt.ylim(-0.5, 7)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper left')
plt.title('🔍 Suma de Vectores: Regla del Paralelogramo', fontsize=14, fontweight='bold')
plt.xlabel('Componente X')
plt.ylabel('Componente Y')

# Marcar puntos importantes
plt.plot(0, 0, 'ko', markersize=8, label='Origen')
plt.plot(v1[0], v1[1], 'ro', markersize=8, alpha=0.7)
plt.plot(v2[0], v2[1], 'bo', markersize=8, alpha=0.7)
plt.plot(v3[0], v3[1], 'go', markersize=10, alpha=0.9)

plt.tight_layout()
plt.show()

# === VERIFICACIÓN DE PROPIEDADES ===
print(f"\n✅ === PROPIEDADES VECTORIALES ===")
print(f"Conmutatividad: v1 + v2 = {v1 + v2}, v2 + v1 = {v2 + v1}")
print(f"¿Son iguales? {np.array_equal(v1 + v2, v2 + v1)}")

producto_2d = np.dot(v1, v2)
angulo_2d = np.degrees(np.arccos(producto_2d / (np.linalg.norm(v1) * np.linalg.norm(v2))))
print(f"\nProducto punto: {producto_2d}")
print(f"Ángulo entre vectores: {angulo_2d:.1f}°")
```

# 📊 2. Matrices: Organizando Datos Multidimensionales

## ¿Qué es una matriz?

Una matriz es una **tabla rectangular de números** que organiza múltiples vectores de forma estructurada.

### En data science:
- **Filas** = Observaciones, muestras, ejemplos
- **Columnas** = Características, variables, features

### Ejemplos:
- 🏦 **Dataset bancario**: [clientes × características_financieras]
- 🖼️ **Imagen**: [píxeles_alto × píxeles_ancho]
- 📺 **Netflix**: [usuarios × películas_valoradas]
- 🛒 **E-commerce**: [productos × características]

---

## 🏗️ Creación y Propiedades de Matrices

```python
# === CREAR MATRIZ DE DATOS BANCARIOS ===
print("🏦 === DATASET BANCARIO COMO MATRIZ ===")

# Cada fila = un cliente, cada columna = una característica
# Columnas: [edad, ingresos_anuales, gastos_anuales, score_crediticio, años_cliente]
clientes_matriz = np.array([
    [25, 50000, 30000, 720, 2],   # Cliente joven profesional
    [35, 70000, 45000, 780, 5],   # Cliente establecido
    [28, 55000, 33000, 690, 3],   # Cliente promedio
    [42, 85000, 52000, 820, 8],   # Cliente premium
    [30, 60000, 38000, 750, 4],   # Cliente en crecimiento
    [38, 75000, 48000, 800, 6]    # Cliente consolidado
])

print("Matriz de clientes bancarios:")
print(clientes_matriz)

# === PROPIEDADES DE LA MATRIZ ===
filas, columnas = clientes_matriz.shape
print(f"\n📊 === PROPIEDADES DE LA MATRIZ ===")
print(f"Forma (shape): {clientes_matriz.shape}")
print(f"Número de clientes (filas): {filas}")
print(f"Número de características (columnas): {columnas}")
print(f"Total de elementos: {clientes_matriz.size}")
print(f"Dimensiones: {clientes_matriz.ndim}D")
print(f"Tipo de datos: {clientes_matriz.dtype}")
print(f"Memoria utilizada: {clientes_matriz.nbytes} bytes")

# === CREAR DATAFRAME PARA MEJOR VISUALIZACIÓN ===
caracteristicas = ['Edad', 'Ingresos', 'Gastos', 'Score_Credit', 'Años_Cliente']
indices_clientes = [f'Cliente_{i+1}' for i in range(filas)]

df_clientes = pd.DataFrame(clientes_matriz, 
                          columns=caracteristicas,
                          index=indices_clientes)

print(f"\n📋 === VISUALIZACIÓN COMO DATAFRAME ===")
print(df_clientes)

# === ACCESO A DATOS ===
print(f"\n🔍 === ACCESO A ELEMENTOS ===")
print(f"Primer cliente (fila 0): {clientes_matriz[0]}")
print(f"Todos los ingresos (columna 1): {clientes_matriz[:, 1]}")
print(f"Score del Cliente 3: {clientes_matriz[2, 3]}")
print(f"Primeros 3 clientes:\n{clientes_matriz[:3]}")
print(f"Solo datos financieros (columnas 1-2):\n{clientes_matriz[:, 1:3]}")
```

## 📈 Análisis Estadístico de la Matriz

```python
# === ANÁLISIS ESTADÍSTICO ===
print("📈 === ANÁLISIS ESTADÍSTICO POR CARACTERÍSTICA ===")

for i, caracteristica in enumerate(caracteristicas):
    valores = clientes_matriz[:, i]
    print(f"\n{caracteristica}:")
    print(f"  • Promedio: {np.mean(valores):,.1f}")
    print(f"  • Mediana: {np.median(valores):,.1f}")
    print(f"  • Mínimo: {np.min(valores):,.0f}")
    print(f"  • Máximo: {np.max(valores):,.0f}")
    print(f"  • Desviación estándar: {np.std(valores):,.1f}")
    print(f"  • Rango: {np.max(valores) - np.min(valores):,.0f}")

# === ESTADÍSTICAS GLOBALES ===
print(f"\n🌐 === ESTADÍSTICAS GLOBALES ===")
print(f"Suma total: {np.sum(clientes_matriz):,.0f}")
print(f"Promedio global: {np.mean(clientes_matriz):,.1f}")
print(f"Valor máximo: {np.max(clientes_matriz):,.0f}")
print(f"Valor mínimo: {np.min(clientes_matriz):,.0f}")

# === ENCONTRAR POSICIONES EXTREMAS ===
pos_max = np.unravel_index(np.argmax(clientes_matriz), clientes_matriz.shape)
pos_min = np.unravel_index(np.argmin(clientes_matriz), clientes_matriz.shape)

print(f"\n🎯 === VALORES EXTREMOS ===")
print(f"Valor máximo {np.max(clientes_matriz)} en: {indices_clientes[pos_max[0]]}, {caracteristicas[pos_max[1]]}")
print(f"Valor mínimo {np.min(clientes_matriz)} en: {indices_clientes[pos_min[0]]}, {caracteristicas[pos_min[1]]}")

# === CORRELACIONES SIMPLES ===
print(f"\n🔗 === ANÁLISIS DE CORRELACIÓN ===")
# Correlación entre ingresos y gastos
ingresos = clientes_matriz[:, 1]
gastos = clientes_matriz[:, 2]
correlacion = np.corrcoef(ingresos, gastos)[0, 1]

print(f"Correlación Ingresos-Gastos: {correlacion:.3f}")
if correlacion > 0.7:
    print("💡 Correlación fuerte: A mayor ingreso, mayor gasto")
elif correlacion > 0.3:
    print("💡 Correlación moderada: Relación parcial entre ingreso y gasto")
else:
    print("💡 Correlación débil: Poca relación entre ingreso y gasto")
```

## ⚙️ Operaciones Básicas con Matrices

```python
# === TRANSPUESTA DE MATRIZ ===
print("🔄 === TRANSPUESTA DE MATRIZ ===")

clientes_T = clientes_matriz.T
print(f"Matriz original: {clientes_matriz.shape} (clientes × características)")
print(f"Matriz transpuesta: {clientes_T.shape} (características × clientes)")
print(f"\nPrimeras 3 columnas de la transpuesta:")
print(clientes_T[:, :3])

# === OPERACIONES ELEMENT-WISE ===
print(f"\n➕ === OPERACIONES ELEMENT-WISE ===")

# Simular datos del año anterior (5% menos)
clientes_año_anterior = clientes_matriz * 0.95

# Suma de matrices
suma_matrices = clientes_matriz + clientes_año_anterior
print(f"Suma de matrices (actual + anterior):")
print(f"Cliente 1 actual: {clientes_matriz[0]}")
print(f"Cliente 1 anterior: {clientes_año_anterior[0].astype(int)}")
print(f"Cliente 1 suma: {suma_matrices[0].astype(int)}")

# Diferencia y crecimiento
diferencia_anual = clientes_matriz - clientes_año_anterior
crecimiento_porcentual = (diferencia_anual / clientes_año_anterior) * 100

print(f"\nCrecimiento anual del Cliente 1:")
for i, caracteristica in enumerate(caracteristicas):
    print(f"  • {caracteristica}: {crecimiento_porcentual[0, i]:.1f}%")

# === MULTIPLICACIÓN POR ESCALAR ===
print(f"\n✖️ === MULTIPLICACIÓN POR ESCALAR ===")

# Ajuste por inflación
factor_inflacion = 1.03
clientes_ajustados = clientes_matriz * factor_inflacion

print(f"Ajuste por inflación del 3%:")
print(f"Cliente 1 original: {clientes_matriz[0]}")
print(f"Cliente 1 ajustado: {clientes_ajustados[0].astype(int)}")

# Conversión selectiva (solo columnas monetarias)
tipo_cambio = 0.85
clientes_euros = clientes_matriz.copy().astype(float)
clientes_euros[:, 1:3] *= tipo_cambio  # Solo ingresos y gastos

print(f"\nConversión a euros (solo valores monetarios):")
print(f"Cliente 1 USD: {clientes_matriz[0]}")
print(f"Cliente 1 EUR: {clientes_euros[0].astype(int)}")
```

## 🔍 Indexing y Filtrado Avanzado

```python
# === FILTRADO DE DATOS ===
print("🔍 === FILTRADO AVANZADO DE DATOS ===")

# Filtrar por edad
jovenes = clientes_matriz[clientes_matriz[:, 0] < 30]
maduros = clientes_matriz[clientes_matriz[:, 0] >= 35]

print(f"Clientes jóvenes (< 30 años): {len(jovenes)} encontrados")
print(jovenes)

print(f"\nClientes maduros (>= 35 años): {len(maduros)} encontrados")
print(maduros)

# Filtrar por múltiples condiciones
alta_calidad = clientes_matriz[(clientes_matriz[:, 1] > 60000) & 
                               (clientes_matriz[:, 3] > 750)]

print(f"\nClientes premium (ingresos > 60K Y score > 750): {len(alta_calidad)}")
print(alta_calidad)

# Selección de columnas específicas
solo_financiero = clientes_matriz[:, [1, 2, 3]]  # Ingresos, gastos, score
print(f"\nSolo datos financieros:")
print(solo_financiero)

# === AGREGACIONES ===
print(f"\n📊 === AGREGACIONES Y ESTADÍSTICAS ===")

# Estadísticas por filas (clientes)
promedio_por_cliente = np.mean(clientes_matriz, axis=1)
suma_por_cliente = np.sum(clientes_matriz, axis=1)

print(f"Promedio por cliente: {promedio_por_cliente.round(0)}")
print(f"Suma por cliente: {suma_por_cliente.astype(int)}")

# Estadísticas por columnas (características)
promedio_por_caracteristica = np.mean(clientes_matriz, axis=0)
std_por_caracteristica = np.std(clientes_matriz, axis=0)

print(f"\nPromedio por característica: {promedio_por_caracteristica.round(1)}")
print(f"Desviación estándar: {std_por_caracteristica.round(1)}")

# Identificar clientes extremos
cliente_mayor_promedio = np.argmax(promedio_por_cliente)
cliente_menor_promedio = np.argmin(promedio_por_cliente)

print(f"\n🏆 Cliente con mayor promedio: {indices_clientes[cliente_mayor_promedio]}")
print(f"📉 Cliente con menor promedio: {indices_clientes[cliente_menor_promedio]}")
```

## ⚡ Multiplicación de Matrices: El Corazón del ML

```python
# === MULTIPLICACIÓN DE MATRICES ===
print("⚡ === MULTIPLICACIÓN DE MATRICES ===")

# Crear sistema de scoring crediticio
# Cada fila = un modelo diferente, cada columna = peso de una característica
modelos_scoring = np.array([
    [0.1,  0.3, -0.2,  0.4,  0.1],   # Modelo conservador
    [0.05, 0.4, -0.3,  0.3,  0.15],  # Modelo orientado a ingresos
    [0.2,  0.2, -0.1,  0.35, 0.25]   # Modelo que valora experiencia
])

print("🏦 Modelos de scoring crediticio (pesos):")
modelos_df = pd.DataFrame(modelos_scoring, 
                         columns=caracteristicas,
                         index=['Conservador', 'Ingresos', 'Experiencia'])
print(modelos_df)

# Aplicar modelos: Clientes × Modelos^T = Puntuaciones
puntuaciones = np.dot(clientes_matriz, modelos_scoring.T)

print(f"\n📊 === RESULTADOS DEL SCORING ===")
print(f"Forma clientes: {clientes_matriz.shape}")
print(f"Forma modelos^T: {modelos_scoring.T.shape}")
print(f"Forma resultado: {puntuaciones.shape}")

# Crear DataFrame de puntuaciones
puntuaciones_df = pd.DataFrame(puntuaciones,
                              columns=['Conservador', 'Ingresos', 'Experiencia'],
                              index=indices_clientes)

print("\nPuntuaciones por cliente y modelo:")
print(puntuaciones_df.round(0))

# === ANÁLISIS DE RESULTADOS ===
print(f"\n🏆 === ANÁLISIS DE MEJORES CLIENTES ===")
mejor_por_modelo = puntuaciones_df.idxmax()
puntuacion_maxima = puntuaciones_df.max()

for modelo in puntuaciones_df.columns:
    mejor_cliente = mejor_por_modelo[modelo]
    score = puntuacion_maxima[modelo]
    print(f"Modelo {modelo}: {mejor_cliente} (score: {score:.0f})")

# === EJEMPLO DE CÁLCULO MANUAL ===
print(f"\n🔢 === VERIFICACIÓN MANUAL ===")
cliente_1 = clientes_matriz[0]
modelo_conservador = modelos_scoring[0]

calculo_manual = np.sum(cliente_1 * modelo_conservador)
calculo_numpy = puntuaciones[0, 0]

print(f"Cliente 1: {cliente_1}")
print(f"Modelo conservador: {modelo_conservador}")
print(f"Cálculo manual: {calculo_manual:.0f}")
print(f"Resultado NumPy: {calculo_numpy:.0f}")
print(f"¿Son iguales? {np.isclose(calculo_manual, calculo_numpy)}")
```

## 🔄 Matriz Identidad, Inversa y Determinante

```python
# === CONCEPTOS AVANZADOS DE MATRICES ===
print("🔄 === MATRIZ IDENTIDAD ===")

# La matriz identidad es como el "1" para las matrices
I_2 = np.eye(2)  # Identidad 2x2
I_3 = np.eye(3)  # Identidad 3x3

print(f"Matriz identidad 2×2:\n{I_2}")
print(f"\nMatriz identidad 3×3:\n{I_3}")

# Demostrar propiedad: A × I = A
A_ejemplo = np.array([[2, 3], [4, 5]])
resultado = np.dot(A_ejemplo, I_2)

print(f"\n✅ Propiedad A × I = A:")
print(f"A:\n{A_ejemplo}")
print(f"A × I:\n{resultado}")
print(f"¿Son iguales? {np.array_equal(A_ejemplo, resultado)}")

# === DETERMINANTE ===
print(f"\n🎯 === DETERMINANTE ===")

matrices_ejemplo = {
    "Invertible": np.array([[2, 1], [1, 1]]),
    "No invertible": np.array([[2, 4], [1, 2]]),
    "Identidad": I_2,
    "Escalar 3x": 3 * I_2
}

for nombre, matriz in matrices_ejemplo.items():
    det = np.linalg.det(matriz)
    print(f"\n{nombre}:")
    print(f"Matriz:\n{matriz}")
    print(f"Determinante: {det:.3f}")
    
    if abs(det) > 1e-10:
        print(f"  ✅ Invertible (det ≠ 0)")
        print(f"  📏 Factor de escalamiento: {abs(det):.1f}x")
    else:
        print(f"  ❌ No invertible (det = 0)")

# === MATRIZ INVERSA ===
print(f"\n🔄 === MATRIZ INVERSA ===")

A = matrices_ejemplo["Invertible"]
if abs(np.linalg.det(A)) > 1e-10:
    A_inv = np.linalg.inv(A)
    
    print(f"Matriz A:\n{A}")
    print(f"Matriz A⁻¹:\n{A_inv}")
    
    # Verificar A × A⁻¹ = I
    producto = np.dot(A, A_inv)
    print(f"\nA × A⁻¹ (debe ser identidad):\n{np.round(producto, 10)}")
    print(f"¿Es identidad? {np.allclose(producto, I_2)}")

# === APLICACIÓN: RESOLVER SISTEMA DE ECUACIONES ===
print(f"\n🧮 === RESOLVER SISTEMA DE ECUACIONES ===")

# Sistema: 2x + y = 5, x + y = 3
A_sistema = np.array([[2, 1], [1, 1]])
b_sistema = np.array([5, 3])

print(f"Sistema de ecuaciones:")
print(f"2x + y = 5")
print(f"x + y = 3")
print(f"\nEn forma matricial: A × x = b")
print(f"A =\n{A_sistema}")
print(f"b = {b_sistema}")

# Resolver usando la inversa
x_solucion = np.linalg.solve(A_sistema, b_sistema)
print(f"\n✅ Solución: x = {x_solucion}")
print(f"💡 x = {x_solucion[0]:.1f}, y = {x_solucion[1]:.1f}")

# Verificar la solución
verificacion = np.dot(A_sistema, x_solucion)
print(f"Verificación A × x = {verificacion}")
print(f"¿Correcto? {np.allclose(verificacion, b_sistema)}")
```

# 🎯 3. Aplicación Práctica: Regresión Lineal

## El Poder del Álgebra Matricial

Vamos a ver cómo el álgebra matricial convierte un problema complejo (encontrar la mejor línea para un conjunto de datos) en una **operación elegante**.

### La fórmula mágica:
**β = (X^T X)⁻¹ X^T y**

Esta ecuación resuelve de forma exacta el problema de regresión lineal sin iteraciones ni aproximaciones.

---

## 🏠 Ejemplo: Predicción de Precios de Casas

```python
# === GENERAR DATOS REALISTAS ===
print("🏠 === PREDICCIÓN DE PRECIOS DE CASAS ===")

# Generar dataset sintético pero realista
np.random.seed(42)
n_casas = 100

# Tamaños de casas entre 50 y 300 m²
tamaño_casas = np.random.uniform(50, 300, n_casas)

# Precio = 1500€ por m² + intercepto + ruido
precio_real = 20000 + 1500 * tamaño_casas + np.random.normal(0, 15000, n_casas)
precio_real = np.maximum(precio_real, 50000)  # Precios mínimos

print(f"Dataset generado: {n_casas} casas")
print(f"Tamaño promedio: {np.mean(tamaño_casas):.1f} m²")
print(f"Precio promedio: €{np.mean(precio_real):,.0f}")
print(f"\nEjemplos:")
for i in range(5):
    print(f"  Casa {i+1}: {tamaño_casas[i]:.0f} m² → €{precio_real[i]:,.0f}")

# === PREPARAR DATOS PARA REGRESIÓN ===
print(f"\n🔧 === PREPARACIÓN DE DATOS ===")

# Matriz X: columna de unos (intercepto) + tamaños
X = np.column_stack([np.ones(n_casas), tamaño_casas])
y = precio_real

print(f"Matriz X: {X.shape} (casas × [intercepto, tamaño])")
print(f"Vector y: {y.shape} (precios)")
print(f"\nPrimeras 5 filas de X:")
print(X[:5])
print(f"Primeros 5 precios: {y[:5].astype(int)}")
```

## 🧮 Aplicando la Fórmula Matricial

```python
# === APLICAR LA FÓRMULA: β = (X^T X)^(-1) X^T y ===
print("🧮 === REGRESIÓN CON ÁLGEBRA MATRICIAL ===")

# Paso a paso para entender cada operación
print("Aplicando β = (X^T X)^(-1) X^T y paso a paso:")

# Paso 1: X^T
X_T = X.T
print(f"\n1. X^T: {X_T.shape}")

# Paso 2: X^T X
XtX = np.dot(X_T, X)
print(f"2. X^T X: {XtX.shape}")
print(f"   X^T X =\n{XtX}")

# Paso 3: (X^T X)^(-1)
XtX_inv = np.linalg.inv(XtX)
print(f"\n3. (X^T X)^(-1):")
print(f"   {XtX_inv}")

# Paso 4: X^T y
Xty = np.dot(X_T, y)
print(f"\n4. X^T y: {Xty}")

# Paso 5: β = (X^T X)^(-1) X^T y
beta = np.dot(XtX_inv, Xty)
intercepto, pendiente = beta

print(f"\n5. β = (X^T X)^(-1) X^T y:")
print(f"   β = {beta}")
print(f"   📊 Intercepto: €{intercepto:,.0f}")
print(f"   📈 Pendiente: €{pendiente:,.0f} por m²")

# === INTERPRETACIÓN DEL MODELO ===
print(f"\n🏠 === MODELO FINAL ===")
print(f"Ecuación: Precio = {intercepto:,.0f} + {pendiente:,.0f} × Tamaño")
print(f"")
print(f"💡 Significado:")
print(f"   • Precio base: €{intercepto:,.0f}")
print(f"   • Precio por m²: €{pendiente:,.0f}")
print(f"   • Casa de 100 m²: €{intercepto + pendiente * 100:,.0f}")
print(f"   • Casa de 200 m²: €{intercepto + pendiente * 200:,.0f}")
```

## 📊 Evaluación del Modelo

```python
# === EVALUACIÓN DEL MODELO ===
print("📊 === EVALUACIÓN DEL MODELO ===")

# Hacer predicciones
y_predicho = np.dot(X, beta)

# Calcular métricas
errores = y - y_predicho
mse = np.mean(errores**2)  # Error cuadrático medio
rmse = np.sqrt(mse)        # Raíz del error cuadrático medio
mae = np.mean(np.abs(errores))  # Error absoluto medio

# R² (coeficiente de determinación)
ss_res = np.sum(errores**2)  # Suma cuadrados residuales
ss_tot = np.sum((y - np.mean(y))**2)  # Suma cuadrados totales
r2 = 1 - (ss_res / ss_tot)

print(f"Métricas de rendimiento:")
print(f"  • MSE: {mse:,.0f}")
print(f"  • RMSE: €{rmse:,.0f}")
print(f"  • MAE: €{mae:,.0f}")
print(f"  • R²: {r2:.3f} ({r2:.1%})")

# Interpretación del R²
if r2 > 0.9:
    calidad = "🟢 Excelente"
elif r2 > 0.7:
    calidad = "🟡 Bueno"
elif r2 > 0.5:
    calidad = "🟠 Aceptable"
else:
    calidad = "🔴 Pobre"

print(f"\n💡 {calidad}: {r2:.1%} de la variación se explica por el tamaño")

# === COMPARACIÓN CON SKLEARN ===
print(f"\n🔍 === VERIFICACIÓN CON SCIKIT-LEARN ===")

modelo_sklearn = LinearRegression()
modelo_sklearn.fit(tamaño_casas.reshape(-1, 1), y)

print(f"Nuestro cálculo:")
print(f"  • Intercepto: €{intercepto:,.0f}")
print(f"  • Pendiente: €{pendiente:,.0f}")
print(f"  • R²: {r2:.6f}")

print(f"\nScikit-learn:")
print(f"  • Intercepto: €{modelo_sklearn.intercept_:,.0f}")
print(f"  • Pendiente: €{modelo_sklearn.coef_[0]:,.0f}")
print(f"  • R²: {modelo_sklearn.score(tamaño_casas.reshape(-1, 1), y):.6f}")

diferencia_intercepto = abs(intercepto - modelo_sklearn.intercept_)
diferencia_pendiente = abs(pendiente - modelo_sklearn.coef_[0])

print(f"\n✅ Diferencias:")
print(f"  • Intercepto: €{diferencia_intercepto:.2f}")
print(f"  • Pendiente: €{diferencia_pendiente:.2f}")
print(f"\n🎯 ¡Álgebra matricial produce resultados idénticos!")
```

```python
# === VISUALIZACIÓN COMPLETA ===
print("📈 === VISUALIZACIÓN DE LA REGRESIÓN ===")

fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Datos y línea de regresión
ax1 = axes[0, 0]
ax1.scatter(tamaño_casas, precio_real, alpha=0.6, s=50, color='blue', label='Datos reales')

# Línea de regresión
tamaño_linea = np.linspace(50, 300, 100)
precio_linea = intercepto + pendiente * tamaño_linea
ax1.plot(tamaño_linea, precio_linea, 'r-', linewidth=2, 
         label=f'Regresión: y = {intercepto:,.0f} + {pendiente:,.0f}x')

ax1.set_xlabel('Tamaño (m²)')
ax1.set_ylabel('Precio (€)')
ax1.set_title('🏠 Regresión Lineal: Precio vs Tamaño')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Residuos
ax2 = axes[0, 1]
ax2.scatter(y_predicho, errores, alpha=0.6, s=50, color='green')
ax2.axhline(y=0, color='red', linestyle='--', alpha=0.8)
ax2.set_xlabel('Valores Predichos (€)')
ax2.set_ylabel('Residuos (€)')
ax2.set_title('📊 Análisis de Residuos')
ax2.grid(True, alpha=0.3)

# 3. Real vs Predicho
ax3 = axes[1, 0]
ax3.scatter(precio_real, y_predicho, alpha=0.6, s=50, color='purple')
min_precio = min(min(precio_real), min(y_predicho))
max_precio = max(max(precio_real), max(y_predicho))
ax3.plot([min_precio, max_precio], [min_precio, max_precio], 'r--', alpha=0.8, label='Línea perfecta')
ax3.set_xlabel('Precio Real (€)')
ax3.set_ylabel('Precio Predicho (€)')
ax3.set_title(f'🎯 Real vs Predicho (R² = {r2:.3f})')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Distribución de errores
ax4 = axes[1, 1]
ax4.hist(errores, bins=20, alpha=0.7, color='orange', edgecolor='black')
ax4.axvline(x=0, color='red', linestyle='--', alpha=0.8)
ax4.set_xlabel('Error (€)')
ax4.set_ylabel('Frecuencia')
ax4.set_title('📈 Distribución de Errores')
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# === EJEMPLOS DE PREDICCIÓN ===
print(f"\n🏠 === EJEMPLOS DE PREDICCIÓN ===")
tamaños_ejemplo = [75, 100, 150, 200, 250]

print(f"Predicciones para diferentes tamaños:")
for tamaño in tamaños_ejemplo:
    precio_predicho = intercepto + pendiente * tamaño
    print(f"  • {tamaño:3d} m²: €{precio_predicho:,.0f}")

# === ANÁLISIS DE CALIDAD ===
print(f"\n🔍 === ANÁLISIS DE CALIDAD ===")
error_porcentual = np.abs(errores) / precio_real * 100

print(f"Error porcentual promedio: {np.mean(error_porcentual):.1f}%")
print(f"Error porcentual mediano: {np.median(error_porcentual):.1f}%")
print(f"Predicciones dentro del ±10%: {np.mean(error_porcentual <= 10) * 100:.1f}%")
print(f"Predicciones dentro del ±20%: {np.mean(error_porcentual <= 20) * 100:.1f}%")

print(f"\n🎉 ¡Modelo completado con álgebra matricial!")
```
